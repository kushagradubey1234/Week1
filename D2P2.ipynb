```python
#log time execution using decorator
import time

def log_execution_time(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print("Execution Time: ", end-start, "seconds")
        return result
    return wrapper
@log_execution_time
def sample_function():
    time.sleep(2)
sample_function()


```
