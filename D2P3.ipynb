```python
#3. How do you create and raise custom exceptions?
"""Custom exceptions are user defined error type. It is created when the program needs to be stopped aur paused at a certain point but it does not have any built-in errortype.
It is created with any name as per the condition and type of error also it is raised using raise keyword. Raising a custom exception is used we need to stop the execution of program."""

#Example

class InvalidAgeError(Exception):
    def __init__(self, age):
        self.age = age
        super().__init__("Invalid age provided: " + str(age))
age = -5
if age < 0:
    raise InvalidAgeError(age)
```
