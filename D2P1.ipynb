```python
#Create a DataExtractor class with methods to read from CSV and JSON files.
import pandas as pd
import json
import csv

class DataExtractor:
    def from_csv(self, file_path):
        return pd.read_csv(file_path)

    def from_json(self, file_path):
        return pd.read_json(file_path, orient="index").T


extractor = DataExtractor()


data_csv = extractor.from_csv("/content/data.csv")
print(data_csv.head())


data_json = extractor.from_json("/content/sample1.json")
print(data_json)



```
